```python
def ipl_dataExt(file_path: str):
    temp_df = pd.DataFrame()
    mid =1
    record_path=['innings','overs','deliveries']
    meta=[['ininings','team'],['ininings','overs','over'],['info']]
    for file in file_path:
        with open(file) as f:
            data = json.load(f)
        df_2 = pd.json_normalize(data, record_path=record_path,meta=meta, errors='ignore')
        df_2 = df_2[['runs.total', 'wickets','ininings.team','ininings.overs.over','info']]
        df_2['ininings'] = df_2['ininings.team'].apply(lambda x: 1 if x == df_2['ininings.team'].iloc[0]  else 2)
        df_2['wicket'] = df_2['wickets'].fillna(0).apply(lambda x: x if type(x) is int else 1)
        a = df_2['ininings.team'].unique()
        if len(a)>1:
            df_2['team_bowling'] = df_2['ininings'].apply(lambda x: a[1] if x==1 else a[0])
        else:
            df_2['team_bowling'] = None
        df_2[['runs_cumsum', 'wicket_cumsum']] = df_2.groupby(['ininings']).cumsum()
        df_2['city'] = df_2['info'].apply(lambda x: x['city'] if 'city' in x.keys() else None)
        df_2['dates'] = df_2['info'].apply(lambda x: x['dates'][0])
        df_2['season'] = df_2['info'].apply(lambda x: "IPL - "+str(x['season']))
        df_2['winner'] = df_2['info'].apply(lambda x: x['outcome']['winner'] if 'winner'in x['outcome'].keys() else None)
        df_2['Match_id'] = mid
        df_2 = df_2[['Match_id', 'season', 'dates','winner','ininings','ininings.team','team_bowling','runs.total','runs_cumsum','wicket_cumsum','ininings.overs.over','city']]
        mid = mid + 1
        temp_df=temp_df.append(df_2)
    return temp_df
```

```python
IPL_Data_Cleaning_cached_df = ipl_dataExt(ipl_files_path)
```

```python
def t20_dataExt(file_path: str):
    temp_df = pd.DataFrame()
    mid =1
    record_path=['innings','overs','deliveries']
    meta=[['ininings','team'],['ininings','overs','over'],['info']]
    for file in file_path:
        with open(file) as f:
            data = json.load(f)
        df_2 = pd.json_normalize(data, record_path=record_path,meta=meta, errors='ignore')
        if 'wickets' not in df_2.columns:
            df_2.insert(9,'wickets',np.nan)
        df_2 = df_2[['runs.total', 'wickets','ininings.team','ininings.overs.over','info']]
        a = df_2['ininings.team'].unique()
        if len(a) > 1:
            df_2['ininings'] = df_2['ininings.team'].apply(lambda x: 1 if x == df_2['ininings.team'].iloc[0]  else 2)
        else:
            df_2['ininings'] = df_2['ininings.team'].apply(lambda x: df_2['ininings.team'].iloc[0])
        df_2['wicket'] = df_2['wickets'].fillna(0).apply(lambda x: int(x) if (type(x) is int) | (type(x) is float) else 1)
        if len(a)>1:
            df_2['team_bowling'] = df_2['ininings'].apply(lambda x: a[1] if x==1 else a[0])
        else:
            df_2['team_bowling'] = None
#         df_2[['runs_cumsum', 'wicket_cumsum']] = df_2.groupby(['ininings']).cumsum()
        df_2[['runs_cumsum', 'wicket_cumsum']] = df_2.groupby(['ininings'])['runs.total','wicket'].cumsum()
        df_2['city'] = df_2['info'].apply(lambda x: x['city'] if 'city' in x.keys() else None)
        df_2['dates'] = df_2['info'].apply(lambda x: x['dates'][0])
        df_2['winner'] = df_2['info'].apply(lambda x: x['outcome']['winner'] if 'winner'in x['outcome'].keys() else None)
        df_2['Match_id'] = mid
        df_2 = df_2[['Match_id','dates','winner','ininings','ininings.team','team_bowling','runs.total','runs_cumsum','wicket_cumsum','ininings.overs.over','city']]
#         print(file)
        mid = mid + 1
        temp_df=temp_df.append(df_2)
    return temp_df
```

```python
t20_Data_Cleaning_cached_df = t20_dataExt(t20_files_path)
```
